# filter

The filter() method filter the given sequence with the help of a function that testes each elements

```python
def even(x):
    return x%2==0
l1 = [12,41,57,29,24,20,4,9]
filter(even,l1)
```

```python
def even(x):
    return x%2==0
l1 = [12,41,57,29,24,20,4,9]
list(filter(even,l1))
```

```python
def even(x):
    return x%2==0
l1 = [12,41,57,29,24,20,4,9]
set(filter(even,l1))
```

```python
def even(x):
    return x%2==0
l1 = [12,41,57,29,24,20,4,9]
tuple(filter(even,l1))
```

```python
list(filter(lambda x:x%2==0, l1))
```

```python
nums = [-2, 5, -7, 10, 0]
positive = list(filter(lambda x: x > 0, nums))
print(positive)
```

```python
names = ["Asha", "Ravi","Arun","Maa","ME"]
result =list(filter(lambda n: n.startswith("M"), names))
print(result)
```

```python
names = ["ram", "sita","arjun","Maa","ME"]
result =list(map(lambda n: n.upper(),names))
print(result)
```

```python
a = [1,2,3]
b = [4,5,6]

res =list(map(lambda x, y: x+y, a, b))
print(res)
```

# Reduce

The reduce(fun,seq) function is used to apply a particular fun a defined in "functool" modlue

```python
from functools import reduce
```

```python
l2 = [1,2,3,4,5]
```

```python
print(l2)
```

```python
reduce(lambda x,y:x+y, l2)
```

```python
1+2+3+4+5
```

```python
lis = [1,3,5,6,2]
```

```python
print(reduce(lambda a, b: a if a > b else b, lis))
```
