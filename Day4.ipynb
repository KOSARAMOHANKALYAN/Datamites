# %s operator

- %s is a converstion type of string

```python
name = 'ABC'
age = 24
print('my name is %s and my age is %s '%(name,age))
```

```python
number = 3
x = int("%s%s"%(number,number))
print(x)
```

# Typecasting

Type Casting is the process of variable's data type converted into another type
There are 2 ways in which typecasting occurs:
1.Autotypecasting
2.Forced Typecasting

The conversion of one data type into other data type is known as  type casting in python



# Auto Typecasting
 The python interpreter automaticallu converts one data type to another without any user involvement

```python
4 + 5.67 + False
```

```python
4 + 5.6 + False + True
```

```python
4+3+'data' +6.7
```
