# Conditional statements

- if 
- elif 
- else

- The if statement evaluate an expression and excutes a block of code on the condition that the expression is true.
- In case the expression returns a False, the if block-of-code is ignored

```python
temp = 36
if temp > 35:
    print("It is hot") # this place consist of indentation space 
```

Indentation refers to the spaces at the beginning of a code line .it is the standard syntax in control statements where python uses indentation to indicate a block of code

```python
temp = 23
if temp > 35:
    print("It is hot")
```

else may ne included to account for alternate statements when if condition is false

```python
temp = 25
if temp >=35:
    print("It is hot")
else:
    print("It is cold")
```

```python
score = 73
if score >=35:
    print("Passed")
else:
    print("Failed")
```

```python
country = "InDia"
if country==country.upper():
    print("The font entered is ALL CAPITAL")
elif country==country.lower():
    print("The font entered is lowercase")
elif country==country.capitalize():
    print("The font receiverd is apperopriate.")
else:
    print("Mixed case received")
print("One of the 4 blocks of the code was executed.")
```

# Nested if-else

 An if condition can be said to be nested when it is part of another if block 

```python
## Nested if-else
num = int(input("Enter the number: "))
if num >5:
    if num <8:
        print(num + 5)
    else:
        print(num + 2)
else:
    print(num)
```
