# Loops

The process of performing a set of tasks iteratively as per requirement is known as looping


There are two essential loops in python

# Control flow statements

- For
- While

For loop is used when we know how many times we want to repeat.

While loop runs until a condition becomes false.

```python
str1 = 'hello'
for x in str1:
    print(x)
```

```python
str2 = 'mohankalyan'
for me in str2:
    print(me)
```

```python
l1 = ['apple','banana','kiwi','guava']
for i in l1:
    print(i)
```

```python
for i in range(1,10):
    print(i)
```

```python
for i in 'hello':
    print(i)
```

```python
s1 =(1,2,'ABC',True)
for i in s1:
    print(i)
```
